```python
import pandas as pd
titanic_df = pd.read_csv('/content/train.csv')

titanic_df.head(n=5)

rows, columns = titanic_df.shape
print(f'All rows: {rows}, all columns: {columns}')

titanic_df.all()
titanic_df.dropna(inplace=True)
titanic_df.to_csv('cleaned_titanic.csv', index = False)

rows, columns = titanic_df.shape
print(f'All rows with cleaned file: {rows}, all columns with cleaned file: {columns}')

titanic_df.head(n=5)
avg_age = round(titanic_df['Age'].mean())
print(f'Average age of the passengers: {avg_age}')

avg_age_gender = round(titanic_df.groupby('Sex')['Age'].mean())
print(f'Average age by passenger gender: {avg_age_gender}')

survivors = titanic_df['Survived'].sum()
print(f'Total survivors: {survivors}')

titanic_df.head(n=5)
max_age = titanic_df['Age'].max()
print(f'The oldest passenger: {max_age}')

min_age = titanic_df['Age'].min()
print(f'The youngest passenger: {min_age}')

titanic_df.head(n=5)
genders = titanic_df['Sex'].value_counts()
print(f'Female and male passengers: {genders}')
titanic_df['Embarked'].value_counts()

import matplotlib.pyplot as plt
titanic_df.head(n=5)
plt.hist(titanic_df['Age'], color='lightgreen', edgecolor='black', alpha=0.7, label='Passenger age')
plt.xlabel('Age')
plt.ylabel('Age frequency')
plt.title('Titanic Passenger Ages')
avg = plt.axvline(titanic_df['Age'].mean(), linestyle = 'dashed', color = 'blue', label = 'Average age')
min = plt.axvline(titanic_df['Age'].min(), linestyle = 'dashed', color = 'orange', label = 'Youngest')
max = plt.axvline(titanic_df['Age'].max(), linestyle = 'dashed', color = 'red', label = 'Oldest')
plt.legend(handles=[avg, min, max])
```
